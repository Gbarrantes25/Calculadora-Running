```python
# Cdigo
# 1. Definimos una lista con los clculos a realizar.
calculos = ["Ritmo de Carrera (Pace).","Velocidad (Km/h).","Frecuencia Cardiaca Mxima (ppm).","Zonas de Entrenamiento.","Caloras Quemadas (kcal)","Salir"]


# 2. Definimos algunas variables con los mensajes que ms utilizaremos.
consulta = "Desea volver al men principal? [SI/NO]"
mensaje_despedida = "Gracias, esperanmos volverte a ver pronto... "


# 3. Ttulo del Proyecto
print("CALCULADORA RUNNING")


# 4. Creamos una funcin "Menu" donde se solicitar la opcion al usuario y se retornar la entrada (un valor entero).
# Esta funcin ser vital para reutilizar cuando el usuario decida regresar al men principal.
def menu():
    entrada = int(input(f"Ingresa una opcin:\n1. {calculos[0]} \n2. {calculos[1]} \n3. {calculos[2]} \n4. {calculos[3]} \n5. {calculos[4]} \n\n6. {calculos[5]}\n\n"))
    return entrada


# 5. Funcin que validar que la opcin del usuario no supere a la de la lista de clculos.
def validacion(e):
    while True:
        if e < 1 or e > 6:
            e = int(input("Ingresa una opcin vlida"))
        else:
            return e


# 6. Ejecutando validacin con la entrada del usuario.
entrada = validacion(menu())


# 7. El bucle while se evaluar "True" y realizar una operacin en base a la opcin dada por el usuario.
# Cuando se haya ejecutado dicha operacin, el usuario podr decidir si regresar al men principal o salir del programa.
while True:
    if entrada == 1:
        print("Ritmo de Carrera\n")
        print("__________________")
        print("A continuacin debers registrar la distancia (Km) y el tiempo (Horas, minutos y segundos)\nque te tom realizar tu carrera.\n")
        distancia = float(input("Ingresa la distancia en Km: "))
        horas = int(input("Ingresa las horas: "))
        minutos = int(input("Ingresa los minutos: "))
        segundos = int(input("Ingresa los segundos: "))
        conversion = ((horas*60) + minutos + (segundos/60))/distancia
        print("\n")
        print(f"Tu ritmo de carrera es: {int(conversion)}:{round((conversion-int(conversion))*60)} Km/min \n")
        pregunta = input(f"{consulta}").lower()
        if pregunta == "si":
            entrada = validacion(menu())
        else:
            print(mensaje_despedida)
            break
            
    
    elif entrada == 2:
        print("Velocidad Km/h")
        print("_______________")
        print("A continuacin debers registrar la distancia (Km) y el tiempo (Horas, minutos y segundos)\nque te tom realizar tu carrera.\n")
        distancia = float(input("Ingresa la distancia en Km: "))
        horas = int(input("Ingresa las horas: "))
        minutos = int(input("Ingresa los minutos: "))
        segundos = int(input("Ingresa los segundos: "))
        conversion = distancia / (horas + (minutos/60) + (segundos/3600))
        print("\n")
        print(f"Tu velocidad de carrera es: {round(conversion,2)} Km/h \n")
        pregunta = input(f"{consulta}").lower()
        if pregunta == "si":
            entrada = validacion(menu())
        else:
            print(mensaje_despedida)
            break

    
    elif entrada == 3:
        print("Frecuencia Cardiaca Mxima")
        print("_______________")
        print("A continuacin debers registrar tu edad.\n")
        edad = int(input("Ingresa tu edad: "))
        conversion = 208 - (edad*0.7)
        print("\n")
        print(f"Tu frecuencia cardiaca mxima es: {round(conversion,2)} ppm \n")
        pregunta = input(f"{consulta}").lower()
        if pregunta == "si":
            entrada = validacion(menu())
        else:
            print(mensaje_despedida)
            break

    
    elif entrada == 4:
        print("Zonas de Entrenamiento")
        print("_______________")
        print("A continuacin debers registrar tu frecuencia cardiaca mxima (FCM).\n")
        fcm = float(input("Ingresa tu frecuencia cardiaca mxima: "))
        zona1 = round(fcm * 0.5,2)
        zona2 = round(fcm * 0.6,2)
        zona3 = round(fcm * 0.7,2)
        zona4 = round(fcm * 0.8,2)
        zona5 = round(fcm * 0.9,2)
        print("\n")
        print(f"Zona 1 (Recuperacin): Entre {zona1} y {zona2} ppm")
        print(f"Zona 2 (Aerbico): Entre {zona2} y {zona3} ppm")
        print(f"Zona 3 (Umbral): Entre {zona3} y {zona4} ppm")
        print(f"Zona 4 (Anaerbico): Entre {zona4} y {zona5} ppm")
        print(f"Zona 5 (Intenso): Entre {zona5} a ms ppm\n")
        pregunta = input(f"{consulta}").lower()
        if pregunta == "si":
            entrada = validacion(menu())
        else:
            print(mensaje_despedida)
            break


    elif entrada == 5:
        print("Caloras Quemadas")
        print("_______________")
        print("A continuacin debers registrar tu peso (Kg) y distancia recorrida (Km).\n")
        peso = float(input("Ingresa tu peso en kilogramos: "))
        distancia = float(input("Ingresa la distancia en Km: "))
        conversion = round(peso * distancia * 1.036,2)
        print("\n")
        print(f"Quemaste {round(conversion,2)} kcal \n")
        pregunta = input(f"{consulta}").lower()
        if pregunta == "si":
            entrada = validacion(menu())
        else:
            print(mensaje_despedida)
            break
    
    elif entrada == 6:
        print(mensaje_despedida)
        break

    
    else:
        break
```
